```python
def filter_candidates(candidates, min_stars):
    black_list = ["Microsoft","Meta Research", "Google", "BCG Gamma", "BCG-Gamma","Amazon", "IBM", "salesforce",
                 "facebookresearch", "Netflix"
                 ]
    retained_langugages = ["Python", "C++", "R", "Java", "TypeScript", "Julia", "Scala", "MATLAB", "C", ]

    regex_black_list = "*|".join(black_list) + "*"
    mask = (
        (candidates["repo_stargazers_count"] > 5) 
        & ~(candidates.orga_name.str.contains(regex_black_list, na=False))
        & ~candidates.repo_language.isnull()
        & (candidates.repo_language.isin(retained_langugages))
    )
    return candidates.assign(**{"to_keep": np.where(mask, 1, 0)})
repositories = filter_candidates(repositories, min_stars=5)
```

```python
print(repositories[repositories.to_keep == 1].groupby(["query"]).repo_id.nunique().sort_values(ascending=False))
print()
print(repositories[repositories.to_keep == 1].groupby(["query"]).repo_stargazers_count.median().sort_values(ascending=False))
print()
print(repositories[repositories.to_keep == 1].groupby(["query"]).repo_stargazers_count.mean().sort_values(ascending=False))
```
